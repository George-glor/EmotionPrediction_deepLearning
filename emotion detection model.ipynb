```python
import cv2
import numpy as np
from keras.models import load_model
from keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt
```

```python
# Load the pre-trained emotion detection model
try:
    emotion_model = load_model('emotion_detection_model.keras')
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")
```

```python
video_capture = cv2.VideoCapture(0)
if not video_capture.isOpened():
    print("Error: Could not open webcam.")
    exit()
else:
    print("Webcam initialized successfully.")
```

```python
# Example of plotting results
emotions = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
predicted_emotion = [0.1, 0.05, 0.15, 0.6, 0.05, 0.03, 0.02]  # Just example data

plt.bar(emotions, predicted_emotion)
plt.xlabel("Emotions")
plt.ylabel("Prediction Probability")
plt.title("Emotion Prediction")
plt.show()
```
